```python
names = df['Name'].unique()

# 随机生成 Salary（在 3000 ~ 10000 之间）
salary = np.random.randint(3000, 10000, size=len(names))

# 创建 df_salary
df_salary = pd.DataFrame({'Name': names, 'Salary': salary})

# 合并 df_salary 和 df
df_merge = pd.merge(df, df_salary, on='Name', how='left')
df_merge['Salary'] = df_merge[['Salary_x', 'Salary_y']].max(axis=1)  # 选择非空的值
df_merge.drop(columns=['Salary_x', 'Salary_y'], inplace=True)  # 删除多余列
# 查看结果
df_merge.head()
df_bonus = df_salary.copy()
df_bonus['Bonus'] = df_bonus['Salary'] * 0.1

# 合并到 df_merge
df_merge = pd.merge(df_merge, df_bonus[['Name', 'Bonus']], on='Name', how='left')

# 计算总收入
df_merge['Total_Earnings'] = df_merge['Salary'] + df_merge['Bonus']

# 找出收入最高的 3 位学生
top_3_earners = df_merge.nlargest(3, 'Total_Earnings')

# 输出结果
top_3_earners

```
