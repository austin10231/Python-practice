```python
def temperature(temp):
    if temp > 40:
        message = "too hot"
    elif temp > 35:
        message = "normal"
    else:
        message = "cold"
    return message
print(temperature(44))
```

```python
def check_bmi(BMI):
    if BMI > 30:
        message = "obese"
    elif 25 < BMI <= 30:
        message = "overweight"
    elif 18.5 < BMI <= 25:
        message = "normal"
    else:
        message = "underweight"
    return message

print(check_bmi(32))     # 输出：obese
print(check_bmi(26.5))   # 输出：overweight
print(check_bmi(22.3))   # 输出：normal
print(check_bmi(17.8))   # 输出：underweight
```

```python
def get_tax(salary):
    if salary < 12000:
        tax = 0.25 * salary
    else:
        tax = 0.3 * salary
    return tax


print(get_tax(10000))
print(get_tax(20000))
```
